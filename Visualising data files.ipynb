```python
# How to set index position as date
df = pd.read_csv(r"C:\Users\user\Desktop\Thokozani\Thokozani\panda  data\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
#How to manipulate data by drawing graphs
df.plot()
```

```python
#How to customize a line chart
#Get the kind of graph e.g line the original one
df.plot(kind = 'line')
```

```python
# By default you get line graph
# How to sub plot columns for charts 
#where every column will have its own chart
df.plot(kind = 'line', subplots = True)
```

```python
# How to visualise all columns 
#add title for a chart
df.plot(kind = 'line', title = 'Ice Cream Rating')
```

```python
# How to customize X and Y axis labels
# X= for datelabel(daily ratings)
# Y= for column heading label(scores)
df.plot(kind = 'line', title = 'Ice Cream Rating', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
# How to change type of graphs or creating a bar graph
df.plot(kind ='bar')
```

```python
# How to change type of graphs or creating a bar graph
#How to make stacked bar chart
df.plot(kind ='bar', stacked = True)
```

```python
#How to target one column chart e.g flavor rating
df['Flavor Rating'].plot(kind ='bar', stacked = True)
```

```python
#How to make a horizontal stacked bar chart for all columns
df.plot.barh(stacked = True)
```

```python
# How to create a scatter graph
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 550, c = 'Red')
```

```python
# How to create histograms
df.plot.hist()
```

```python
# How to create histograms
# How to specify bins to 20
df.plot.hist(bins = 20)
```

```python
# How to make a box plot graph
df.boxplot()
```

```python
# How to create area chart
df.plot.area()
```

```python
# How to specify area chart size
df.plot.area(figsize = (10,5))
```
